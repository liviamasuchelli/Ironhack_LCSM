```python
df = pd.read_csv('global-shark-attacks.zip',encoding='latin1',compression='zip')
```

```python
df.shape
```

```python
colunas = list(df.columns)
colunas
```

```python
colunas = [nomes.lower() for nomes in colunas]
colunas = [nomes.strip() for nomes in colunas]
colunas
```

```python
df.columns = colunas
df.head(1)
```

```python
df.info()
```

```python
df.dropna(axis = 1, thresh = (0.7 * (df.shape[1])), inplace = True)
df
```

## Dropna em colunas com 70% de missing eliminou as colunas unnamed 22 e unnamed 23

```python
df.dropna(axis = 0, thresh = (0.7 * (df.shape[1])), inplace = True)
df
```
