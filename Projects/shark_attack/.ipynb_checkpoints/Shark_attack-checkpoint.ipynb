# Shark Attacks Data Cleaning

## Steps do Data Cleaning:

1) Normalizar os títulos das colunas
2) Avaliar remoção de colunas, considerando:
    a. Quantidade de dados ausentes
    b. Redundância de conteúdos entre as colunas
3) Considerar a criação de novas colunas para agrupamento e split de dados, facilitando análises posteriores
4) Ao longo do exercício, eliminar conteúdos incompletos.
   

```python
import pandas as pd
```

```python
df = pd.read_csv('global-shark-attacks.zip',encoding='latin1',compression='zip')
```

### 1) Normalizar os títulos das colunas

```python
nomes_colunas = df.columns
nomes_colunas
```

#### Renomeando colunas com problemas no título
##### (espaços à esquerda e direita e letra minúscula)

```python
nomes_colunas = list(nomes_colunas)

nomes_colunas = [name.strip() for name in nomes_colunas]

```

```python
nomes_colunas = [name.capitalize() for name in nomes_colunas]

nomes_colunas
```

```python
df.columns = nomes_colunas 
```

### 2) Avaliar remoção de colunas

##### *Agrupadores para 'Year' que facilitem análises posteriores*

```python
df[df['Year']==0]['Year'].value_counts()
```

```python
df[df['Year']<1900][['Year', 'Type', 'Country']]
```

```python
def Year_group (row):
    if row['Year']<1900:
        return 'Before 1900s'
    if row ['Year']>=1900:
        return 'To be analyzed'
    return 'Check formula Year_Group'
```

```python
df['Year Group'] = df.apply(lambda row : Year_group (row), axis =1 )
```

```python
df.head(2)
```

```python
sub_check_formula = df.loc[df['Year Group'] =='Check formula Year_Group']
sub_check_formula.info()

# Avaliando os casos em que a coluna original "Year" é nula, constata-se que outras colunas também são pobres em dados.

# CONCLUSÃO: FAZ SENTIDO DROPAR AS LINHAS EM QUE 'CHECK FORMULA' APARECE
```
